```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

# Number of points per cluster
n = 100

# Blue and Yellow differ ONLY in z
blue = np.random.randn(n, 3) * 0.4 + np.array([0, 0, -5])
yellow = np.random.randn(n, 3) * 0.4 + np.array([0, 0, 5])

# Green cluster separated in x/y
green = np.random.randn(n, 3) * 0.4 + np.array([5, 5, 0])

# --- 3D plot ---
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

ax.scatter(blue[:, 0], blue[:, 1], blue[:, 2], label="Blue", alpha=0.7)
ax.scatter(yellow[:, 0], yellow[:, 1], yellow[:, 2], label="Yellow", alpha=0.7)
ax.scatter(green[:, 0], green[:, 1], green[:, 2], label="Green", alpha=0.7)

ax.set_title("Original 3D Space")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.show()
```

```python

# --- Project onto x-y plane (drop z) ---
blue_xy = blue[:, :2]
yellow_xy = yellow[:, :2]
green_xy = green[:, :2]

# --- 2D plot ---
plt.figure()
plt.scatter(blue_xy[:, 0], blue_xy[:, 1], label="Blue", alpha=0.7)
plt.scatter(yellow_xy[:, 0], yellow_xy[:, 1], label="Yellow", alpha=0.7)
plt.scatter(green_xy[:, 0], green_xy[:, 1], label="Green", alpha=0.7)

plt.title("Lower dimensional representation")
plt.xlabel("x")
plt.ylabel("y")
plt.axis("equal")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

# Create original data manifold (2D spiral)
n_points = 500
theta = np.linspace(0, 4 * np.pi, n_points)
r = theta
x0 = np.vstack((r * np.cos(theta), r * np.sin(theta))).T

plt.figure()
plt.scatter(x0[:, 0], x0[:, 1], s=5)
plt.title("Original Manifold $Z_0$", fontsize=40)
plt.axis("equal")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

# Recreate manifold
n_points = 500
theta = np.linspace(0, 4 * np.pi, n_points)
r = theta
x0 = np.vstack((r * np.cos(theta), r * np.sin(theta))).T

# Noise
noise = np.random.normal(0, 1, x0.shape)

# Intermediate timestep
alpha_t = 0.5
x_t = np.sqrt(alpha_t) * x0 + np.sqrt(1 - alpha_t) * noise

plt.figure()
plt.scatter(x_t[:, 0], x_t[:, 1], s=5)
plt.title("Intermediate Step $z_t$", fontsize=40)
plt.axis("equal")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

# Pure Gaussian noise
n_points = 500
x_T = np.random.normal(0, 1, (n_points, 2))

plt.figure()
plt.scatter(x_T[:, 0], x_T[:, 1], s=5)
plt.title("Pure Noise $Z_T$", fontsize=40)
plt.axis("equal")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Reproducibility
np.random.seed(42)

# Number of cells per cluster
n_cells = 500

# Nearly identical cluster centers (highly overlapping)
mean_cluster_1 = [0.0, 0.0]
mean_cluster_2 = [20, -20]  # very small shift

# Shared covariance (high variance = homogeneity)
covariance = [[1.2, 0.9],
              [0.9, 1.2]]

# Generate data
cluster_1 = np.random.multivariate_normal(
    mean_cluster_1, covariance, n_cells
)
cluster_2 = np.random.multivariate_normal(
    mean_cluster_2, covariance, n_cells
)

# Plot
plt.figure(figsize=(6, 6))
plt.scatter(cluster_1[:, 0], cluster_1[:, 1],
            s=15, alpha=0.6, label="Cluster 1")
plt.scatter(cluster_2[:, 0], cluster_2[:, 1],
            s=15, alpha=0.6, label="Cluster 2")

plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.title("Highly Homogeneous scRNA-like 2D Clusters")
plt.legend()
plt.tight_layout()
plt.show()
```
